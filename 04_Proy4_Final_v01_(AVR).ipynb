En este punto, se tomarán los procesos de identificación de "outliers" de la variable precio resultados del trabajo dos [4] y se aplicarán las rutinas para imputar los valores "faltantes" de las tres etiquetas identificadas.

**1.2 - ELIMINAR OUTLIERS**

```python
#Se define un rango intercuartil para incluir los datos entre el 25% y el 75% de los datos.
q1, q3 = datos_limpios['price'].quantile(0.25), datos_limpios['price'].quantile(0.75)
IQR = q3 - q1

upper = q3 + 1.5 * IQR
lower = q1 - 1.5 * IQR 

data_r = datos_limpios[(datos_limpios.price > lower) & (datos_limpios.price < upper)]
```

```python
#se aplica un reemplazo de variables para ajustar la escala 
precio = data_r.price/1000

#se grafica nuevamente separando por categoría de propiedad
plt.figure(figsize=(18,10))
sns.boxplot(y = precio, x = data_r.property_type)
```

```python
#Se determinan los estadísticos más relevantes que describen la variable precio con el DataFrame transformado:
data_r["price"].describe()
```

**1.2.1 -INTERPRETACION**
La nueva base de datos sin outliers [data_r], exhibe un precio promedio de 182.280 USD/Propiedad y una desviación de 98.560USD/propiedad, se muestra que
el 75% de los datos está a menos de 1 desviación y el valor extremo está a menos de 6 desviaciones. En comparación, el DF original analizado en el trabajo 2 [4], se notaba un precio promedio de  241.000USD/propiedad, una desviación estándar de 318.000 usd/propiedad, una concentración del 75%  de los datos  por debajo de 1 desviación estándar  y datos Outliers a más de 10 desviaciones estándar del promedio. En conclusión la rutina de  eliminación de Outliers ha sido efectiva.

**1.3 - IMPUTAR VALORES ** 
En este paso, se toman las tres etiquetas con datos faltantes: "Surface_covered", "lon"y "lat" y se aplicará una rutina para identificar el tipo de faltante por etiqueta y poesteriormente un algoritmo para imputar tales espacios faltantes.

```python
#Contar datos faltantes por tipo de propiedad agrupados por "Surface_covered"
sc = data_r.surface_covered.isna().groupby([data_r["property_type"]]).sum().astype(int)
sc
```

La mayor cantidad de valores faltantes en valores absolutos corresponde a las categorias Departamento, Casa y Lote.

```python
#Contar datos faltantes por tipo de propiedad agrupados por "lon"
sc = data_r.lon.isna().groupby([data_r["property_type"]]).sum().astype(int)
sc
```

La mayor cantidad de valores faltantes en valores absolutos corresponde a las categorias Departamento, Casa y Lote.

```python
#Contar datos faltantes por tipo de propiedad agrupados por "lat"
sc = data_r.lat.isna().groupby([data_r["property_type"]]).sum().astype(int)
sc
```

** 1.3.1 INTERPRETACION**
La mayor cantidad de valores faltantes en valores absolutos corresponde a las categorias Departamento, Casa y Lote.
Nótese que para "lat" y "lon" la cantidad de faltantes por tipo de propiedad difiere mínimamente en la la categoría "Departamento" y "Casa" y es identica en "lote", esto puede indicar que en efecto ha sido un dato faltante MAR  (missing at random), pués si no se tiene el valor de latitud es muy probable que no se tenga el de longitud, pués ambos valores se requieren para determinar las coordenadas de un lugar.

```python
#Comprobar los valores MAR en el Dataframe
lo = data_r[(properati_ds.property_type.isin(["Departamento"]))]
lo.sample(20)
```

** 1.3.2 INTERPRETACION**
Nótese que  indagando por la categoría departamento, los valores faltantes por Lon y Lat van  en pares, por lo que corrobora la hipótesis de faltantes tipo MAR.
Para el proceso de imputar los faltantes se tomarán los valores promedio por tipo de propiedad para la etiqueta "surface_covered", ya que es lógico pensar que los tipos de propiedad comparten atributos similares.  Para las etiquetas "lon" y "lat" se tomarán valores promedios de latitud y longitud por cada barrio o localidad de la etiqueta "l3", ya que las coordenadas promedio por barrio podría ser equiparado con el centroide o posición geográfica media de cada zona. Además la etiqueta l3 no presenta valores faltantes en el dataframe.

```python
#Imputar valores faltantes en surface_covered según tipo de propiedad
data_r["surface_covered"] = data_r.groupby("property_type")["surface_covered"].transform(lambda grp: grp.fillna(np.mean(grp)))


#Imputar valores faltantes en bathrooms según tipo de propiedad
data_r["lat"] = data_r.groupby("l3")["lat"].transform(lambda grp: grp.fillna(np.mean(grp)))

#Imputar valores faltantes en surface_total según tipo de propiedad
data_r["lon"] = data_r.groupby("l3")["lon"].transform(lambda grp: grp.fillna(np.mean(grp)))

data_r.isna().sum()
```

```python
#Se procede  a verificar si algunos faltantes han sido imputados, con una muestra de 20 registros
data_r.sample(20)
```

Después de aplicar un algoritmo de imputación sobre los MAR de las etiquetas "surface_covered", "lat" y "lon", se obtiene un dataframe con todos sus entradas ocupadas por valores no-vacios o faltantes.

**1.4 - ENCODING**
El próximo paso consiste en transformar los valores categóricos de las etiquetas en valores numéricos, que permita incorporarlos al algoritmo de ML y realizar una  predicción generalista. Hay dos formas de lograr esto: one hot encoding y label encoding.

En One Hot Encoding vamos a tomar la columna que tiene datos categóricos, que ha sido codificada con etiquetas, y luego dividirla en múltiples columnas. Los números se reemplazan por 1 y 0, dependiendo de qué columna tiene qué valor. El problema, sin embargo, es que si tenemos muchas categorías vamos a terminar con un dataset muy grande.

Label encoding, por su parte, identifica los distintos valores existentes y sustituye cada uno de ellos por un número. El problema que presenta esta codificacion es que, ya que hay diferentes números en la misma columna, el modelo puede malinterpretar los datos para que estén en algún tipo de orden, 0 < 1 < 2, aunque este no sea el caso.
